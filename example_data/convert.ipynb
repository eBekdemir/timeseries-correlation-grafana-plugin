```python
import pandas
```

```python
aapl = pandas.read_csv('aapl.csv')
msft = pandas.read_csv('msft.csv')
aapl.head()
```

```python
aapl['time'] = pandas.to_datetime(aapl['Date']).astype("int64") // 10**6
aapl['AAPL'] = aapl['Close']
msft['time'] = pandas.to_datetime(msft['Date']).astype("int64") // 10**6
msft['MSFT'] = msft['Close']
aapl = aapl[['time', 'AAPL']]
msft = msft[['time', 'MSFT']]
```

```python
aapl.to_csv('aapl_v1.csv', index=False)
msft.to_csv('msft_v1.csv', index=False)
```
